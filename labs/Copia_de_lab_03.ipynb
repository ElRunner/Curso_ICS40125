```python
import pandas as pd

# Cargar datos para asegurar que 'df' esté definido
df = pd.read_csv('/content/netflix_titles.csv')

# Transformar columna 'date_added' a tipo fecha:
df['date_added'] = pd.to_datetime(df['date_added'], format='mixed')

# Crear columnas auxiliares con 'year_added' y 'month_added':
df = df.assign(
    year_added=df['date_added'].dt.year,
    month_added=df['date_added'].dt.month
)

print("\nDataFrame after transformations:\n")
display(df.head())
```

#### 12. ¿Existen películas con el mismo nombre (`title`) pero con distinto año de lanzamiento (`release_year`)?

```python
df_exploded_genres = df.assign(listed_in=df['listed_in'].str.split(', ')).explode('listed_in')

genre_rating_counts = df_exploded_genres.value_counts(subset=['listed_in', 'rating']).nlargest(10)

print("Top 10 combinaciones más frecuentes de género y rating:")
display(genre_rating_counts)
```

```python
duplicate_titles_diff_year = df.groupby('title')['release_year'].nunique()
duplicate_titles_diff_year = duplicate_titles_diff_year[duplicate_titles_diff_year > 1]

if not duplicate_titles_diff_year.empty:
    print("Títulos con el mismo nombre pero distinto año de lanzamiento:")
    # Display the actual movies for these titles
    for title_name in duplicate_titles_diff_year.index:
        display(df.loc[df['title'] == title_name, ['title', 'release_year', 'show_id']])
else:
    print("No hay películas con el mismo nombre pero distinto año de lanzamiento.")
```
