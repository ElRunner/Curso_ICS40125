## 🎨 Parte 2: Tendencias y evolución en el tiempo

4. **Número de títulos agregados por año**

   * Usa `date_added` (convertido a fecha) y grafica una serie temporal.
   * Pregunta guía: ¿Cuándo creció más el catálogo de Netflix? ¿Qué relación podría tener con la expansión internacional de la plataforma?

5. **Heatmap de lanzamientos por año y mes**

   * Construye un mapa de calor con `year_added` y `month_added`.
   * Pregunta guía: ¿Hay meses con más estrenos que otros? ¿Qué conclusiones puedes sacar sobre la estacionalidad de lanzamientos?

6. **Duración de películas por género**

   * Extrae la duración en minutos y crea un boxplot agrupado por género principal.
   * Pregunta guía: ¿Qué géneros suelen durar más? ¿Qué podrías decir sobre la variabilidad dentro de un mismo género?


```python
import matplotlib.pyplot as plt
import seaborn as sns

# 4. Número de títulos agregados por año

# Convertir 'date_added' a formato de fecha y extraer el año
df_copy = df.copy() # Trabajar en una copia para evitar SettingWithCopyWarning
df_copy['date_added'] = pd.to_datetime(df_copy['date_added'], errors='coerce')
df_copy = df_copy.dropna(subset=['date_added']) # Eliminar filas con fechas nulas si las hay
df_copy['year_added'] = df_copy['date_added'].dt.year

# Contar títulos por año
titles_per_year = df_copy['year_added'].value_counts().sort_index()

# Graficar serie temporal
plt.figure(figsize=(12, 6))
sns.lineplot(x=titles_per_year.index, y=titles_per_year.values)
plt.title('Número de Títulos Agregados a Netflix por Año')
plt.xlabel('Año de Incorporación')
plt.ylabel('Número de Títulos')
plt.grid(True)
plt.show()

# Pregunta guía: ¿Cuándo creció más el catálogo de Netflix? ¿Qué relación podría tener con la expansión internacional de la plataforma?
# Para responder a esta pregunta, se debe observar el pico de la gráfica de línea. Generalmente, un aumento significativo en la cantidad de títulos
# podría correlacionarse con períodos de expansión de la plataforma a nuevos mercados internacionales o un aumento en la inversión en contenido original.
```

**¿Cuándo creció más el catálogo de Netflix? ¿Qué relación podría tener con la expansión internacional de la plataforma?**

Para responder a esta pregunta, se debe observar el pico de la gráfica de línea. Generalmente, un aumento significativo en la cantidad de títulos podría correlacionarse con períodos de expansión de la plataforma a nuevos mercados internacionales o un aumento en la inversión en contenido original.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# 5. Heatmap de lanzamientos por año y mes

# Asegurarse de que 'date_added' sea de tipo datetime y 'year_added' exista
# (ya se hizo en el paso anterior, pero se repite por si se ejecuta de forma independiente)
df_copy = df.copy()
df_copy['date_added'] = pd.to_datetime(df_copy['date_added'], errors='coerce')
df_copy = df_copy.dropna(subset=['date_added'])
df_copy['year_added'] = df_copy['date_added'].dt.year
df_copy['month_added'] = df_copy['date_added'].dt.month_name() # Obtener el nombre del mes

# Definir el orden de los meses para una visualización correcta
month_order = ['January', 'February', 'March', 'April', 'May', 'June', 'July',
               'August', 'September', 'October', 'November', 'December']

# Crear una tabla pivote para contar títulos por año y mes
heatmap_data = df_copy.groupby(['year_added', 'month_added']).size().unstack(fill_value=0)
heatmap_data = heatmap_data[month_order] # Reordenar columnas por mes

plt.figure(figsize=(15, 8))
sns.heatmap(heatmap_data, cmap='YlGnBu', fmt='d', linewidths=.5, cbar_kws={'label': 'Número de Títulos Agregados'})
plt.title('Distribución de Títulos Agregados a Netflix por Año y Mes')
plt.xlabel('Mes de Incorporación')
plt.ylabel('Año de Incorporación')
plt.show()

# Pregunta guía: ¿Hay meses con más estrenos que otros? ¿Qué conclusiones puedes sacar sobre la estacionalidad de lanzamientos?
# Para responder a esto, se debe observar las celdas más claras o con números más altos en el heatmap. Es común ver
# una concentración de lanzamientos en ciertos meses, posiblemente coincidiendo con estrategias de marketing o períodos
# festivos para captar la atención de la audiencia.
```

**¿Hay meses con más estrenos que otros? ¿Qué conclusiones puedes sacar sobre la estacionalidad de lanzamientos?**

Para responder a esto, se debe observar las celdas más claras o con números más altos en el heatmap. Es común ver una concentración de lanzamientos en ciertos meses, posiblemente coincidiendo con estrategias de marketing o períodos festivos para captar la atención de la audiencia.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# 6. Duración de películas por género

# Filtrar solo películas
movies_df = df[df['type'] == 'Movie'].copy()

# Extraer la duración en minutos y convertir a numérico
# Primero, reemplazar ' min' y luego convertir a numérico, forzando errores a NaN
movies_df['duration_minutes'] = movies_df['duration'].str.replace(' min', '', regex=False)
movies_df['duration_minutes'] = pd.to_numeric(movies_df['duration_minutes'], errors='coerce')

# Eliminar filas donde 'duration_minutes' sea NaN después de la conversión
movies_df = movies_df.dropna(subset=['duration_minutes'])

# Asegurarse de que la columna sea de tipo entero
movies_df['duration_minutes'] = movies_df['duration_minutes'].astype(int)

# Extraer el género principal (el primero de la lista)
movies_df['primary_genre'] = movies_df['listed_in'].apply(lambda x: x.split(',')[0].strip())

# Contar cuántas películas hay por género para filtrar los que tienen muy pocas muestras
genre_counts = movies_df['primary_genre'].value_counts()
main_genres = genre_counts[genre_counts >= 50].index # Considerar géneros con al menos 50 películas

movies_filtered_by_genre = movies_df[movies_df['primary_genre'].isin(main_genres)]

# Crear el boxplot
plt.figure(figsize=(15, 8))
sns.boxplot(data=movies_filtered_by_genre, x='primary_genre', y='duration_minutes', palette='tab10')
plt.title('Duración de Películas por Género Principal en Netflix')
plt.xlabel('Género Principal')
plt.ylabel('Duración (minutos)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# Pregunta guía: ¿Qué géneros suelen durar más? ¿Qué podrías decir sobre la variabilidad dentro de un mismo género?
# Observando el boxplot, los géneros con las cajas más altas y medianas más elevadas serán los que duren más.
# La longitud de las cajas y la extensión de los 'bigotes' indican la variabilidad. Un rango intercuartílico (IQR) grande
# y bigotes largos sugieren una mayor variabilidad en la duración de las películas dentro de ese género.
```

**¿Qué géneros suelen durar más? ¿Qué podrías decir sobre la variabilidad dentro de un mismo género?**

Observando el boxplot, los géneros con las cajas más altas y medianas más elevadas serán los que duren más. La longitud de las cajas y la extensión de los 'bigotes' indican la variabilidad. Un rango intercuartílico (IQR) grande y bigotes largos sugieren una mayor variabilidad en la duración de las películas dentro de ese género.

   * Usa un gráfico de barras para mostrar los países más frecuentes en la columna `country`.

8. **Películas vs Series según género**

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# 7. Top 10 países con más producciones

# Manejar valores nulos y múltiples países en la columna 'country'
# Primero, llenar NaN con una cadena vacía para poder usar .str.split()
countries_df = df.dropna(subset=['country']).copy() # Eliminar filas donde country es NaN para el análisis
countries_df['country_list'] = countries_df['country'].apply(lambda x: [c.strip() for c in str(x).split(',')])

# Expandir la lista de países para contar cada aparición individual
expanded_countries = countries_df.explode('country_list')

# Contar la frecuencia de cada país
country_counts = expanded_countries['country_list'].value_counts()

# Seleccionar los top 10 países
top_10_countries = country_counts.head(10)

# Crear el gráfico de barras
plt.figure(figsize=(12, 7))
sns.barplot(x=top_10_countries.index, y=top_10_countries.values, hue=top_10_countries.index, palette='viridis', legend=False)
plt.title('Top 10 Países con Más Producciones en Netflix')
plt.xlabel('País')
plt.ylabel('Número de Títulos')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# Pregunta guía: ¿Qué países dominan el catálogo? ¿Sorprende la distribución o es esperada?
# Para responder a esto, se debe observar las barras más altas. Generalmente, Estados Unidos y la India
# suelen dominar este tipo de distribuciones en plataformas globales como Netflix debido a sus grandes industrias
# cinematográficas y de televisión, así como a su audiencia.
```

**¿Qué países dominan el catálogo? ¿Sorprende la distribución o es esperada?**

Para responder a esto, se debe observar las barras más altas. Generalmente, Estados Unidos y la India suelen dominar este tipo de distribuciones en plataformas globales como Netflix debido a sus grandes industrias cinematográficas y de televisión, así como a su audiencia.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# 8. Películas vs Series según género

# Crear una copia para evitar SettingWithCopyWarning
df_genres = df.copy()

# Extraer el género principal para cada título
df_genres['primary_genre'] = df_genres['listed_in'].apply(lambda x: x.split(',')[0].strip())

# Contar la ocurrencia de cada combinación de tipo y género
genre_type_counts = df_genres.groupby(['primary_genre', 'type']).size().unstack(fill_value=0)

# Seleccionar los géneros principales para la visualización (ej. los top N o filtrar por recuento)
# Para este ejemplo, tomaremos los 20 géneros principales para no saturar el gráfico
top_genres = df_genres['primary_genre'].value_counts().head(20).index
genre_type_counts_filtered = genre_type_counts[genre_type_counts.index.isin(top_genres)]

# Asegurarse de que 'Movie' y 'TV Show' estén presentes como columnas, si no, añadirlas con 0s
if 'Movie' not in genre_type_counts_filtered.columns: genre_type_counts_filtered['Movie'] = 0
if 'TV Show' not in genre_type_counts_filtered.columns: genre_type_counts_filtered['TV Show'] = 0

# Ordenar para mejor visualización si es necesario
genre_type_counts_filtered = genre_type_counts_filtered.sort_values(by='Movie', ascending=False)

# Crear el gráfico de barras apiladas
plt.figure(figsize=(16, 9))
genre_type_counts_filtered[['Movie', 'TV Show']].plot(kind='bar', stacked=True, colormap='viridis', ax=plt.gca())
plt.title('Distribución de Títulos (Películas vs. Series) por Género Principal')
plt.xlabel('Género Principal')
plt.ylabel('Cantidad de Títulos')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Tipo de Contenido')
plt.tight_layout()
plt.show()

# Pregunta guía: ¿Existen géneros más asociados a películas o a series?
# Para responder a esto, se debe observar la proporción de 'Movie' vs 'TV Show' dentro de cada barra de género.
# Barras con una gran sección de un color indican una mayor asociación a ese tipo de contenido.
```

**¿Existen géneros más asociados a películas o a series?**

Para responder a esto, se debe observar la proporción de 'Movie' vs 'TV Show' dentro de cada barra de género. Barras con una gran sección de un color indican una mayor asociación a ese tipo de contenido.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# 9. Relación entre duración y año de lanzamiento

# Filtrar solo películas
movies_df_duration = df[df['type'] == 'Movie'].copy()

# Extraer la duración en minutos y convertir a numérico
# Primero, reemplazar ' min' y luego convertir a numérico, forzando errores a NaN
movies_df_duration['duration_minutes'] = movies_df_duration['duration'].str.replace(' min', '', regex=False)
movies_df_duration['duration_minutes'] = pd.to_numeric(movies_df_duration['duration_minutes'], errors='coerce')

# Eliminar filas donde 'duration_minutes' sea NaN después de la conversión
movies_df_duration = movies_df_duration.dropna(subset=['duration_minutes'])

# Asegurarse de que la columna sea de tipo entero
movies_df_duration['duration_minutes'] = movies_df_duration['duration_minutes'].astype(int)

# Crear el gráfico de dispersión
plt.figure(figsize=(15, 8))
sns.scatterplot(data=movies_df_duration, x='release_year', y='duration_minutes', alpha=0.6)
plt.title('Relación entre Duración de Películas y Año de Lanzamiento')
plt.xlabel('Año de Lanzamiento')
plt.ylabel('Duración (minutos)')
plt.grid(True)
plt.tight_layout()
plt.show()

# Pregunta guía: ¿Han cambiado los estándares de duración de las películas con el tiempo? ¿Se ven outliers interesantes?
# Para responder a esto, se debe observar si hay una tendencia clara en los puntos (aumento, disminución, o estabilidad de la duración promedio).
# Los outliers serán puntos que se desvían significativamente del patrón general, ya sea con duraciones extremadamente cortas o largas para su año.
```

* Encuentra las combinaciones más frecuentes de **género + rating** y represéntalas en un **heatmap o gráfico de burbujas**.
* Pregunta guía: ¿Qué géneros están más dirigidos a un público adulto? ¿Y cuáles a público familiar o infantil?

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Desafío Final: Género + Rating

# Crear una copia para evitar SettingWithCopyWarning
df_challenge = df.copy()

# Extraer el género principal para cada título
df_challenge['primary_genre'] = df_challenge['listed_in'].apply(lambda x: x.split(',')[0].strip())

# Eliminar filas donde 'rating' o 'primary_genre' sean nulos
df_challenge = df_challenge.dropna(subset=['rating', 'primary_genre'])

# Contar la ocurrencia de cada combinación de género y rating
genre_rating_counts = df_challenge.groupby(['primary_genre', 'rating']).size().unstack(fill_value=0)

# Filtrar para considerar solo los géneros principales con un número significativo de títulos
# (ej. top 30 géneros o géneros con más de X títulos en total)
top_genres_for_challenge = df_challenge['primary_genre'].value_counts().head(30).index
genre_rating_counts_filtered = genre_rating_counts[genre_rating_counts.index.isin(top_genres_for_challenge)]

# Filtrar para considerar solo los ratings principales (para no saturar el heatmap)
top_ratings = df_challenge['rating'].value_counts().head(15).index
genre_rating_counts_filtered = genre_rating_counts_filtered[top_ratings]

# Crear el heatmap
plt.figure(figsize=(18, 10))
sns.heatmap(genre_rating_counts_filtered, cmap='YlGnBu', annot=True, fmt='d', linewidths=.5, cbar_kws={'label': 'Número de Títulos'})
plt.title('Combinaciones de Género Principal y Clasificación por Edad en Netflix')
plt.xlabel('Clasificación por Edad')
plt.ylabel('Género Principal')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# Pregunta guía: ¿Qué géneros están más dirigidos a un público adulto? ¿Y cuáles a público familiar o infantil?
# Para responder a esto, se debe observar las celdas con valores altos (colores más intensos) para las combinaciones.
# Los géneros con alta frecuencia en clasificaciones como 'TV-MA' y 'TV-14' se dirigen a un público adulto/adolescente.
# Los géneros con alta frecuencia en clasificaciones como 'TV-Y', 'TV-G', 'G', 'PG', 'TV-Y7', 'TV-PG' se dirigen a un público infantil/familiar.
```
